```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
#movies.head(1)
#credits.head(1)
```

```python
fm = movies.merge(credits,on = "title")
#movies.head(1)
#fm.info()
```

```python
sm = fm[['movie_id',"title",'overview','genres','keywords','cast','crew']]
```

```python
import ast
def conv(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
sm['genres'],sm['keywords'] = sm['genres'].apply(conv),sm['keywords'].apply(conv)
```

```python
import ast
def con_cast(obj):
    L = []
    c=0
    for i in ast.literal_eval(obj):
        if (c<3):
            L.append(i['name'])
            c+=1
        else:
            break
    return L
```

```python
sm['cast'] = sm['cast'].apply(con_cast)
```

```python
import ast
def con_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
sm['crew'] = sm['crew'].apply(con_crew)
```

```python
sm['overview'] = sm['overview'].apply(lambda x: x.split() if type(x) is str else 'empty')
```

```python
sm['#'] = sm["genres"]+sm["crew"]+sm["cast"]+sm["keywords"]
sm["#"].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = "english")
```

```python
elist = sm[["movie_id","title","#"]]

```

```python
elist.head(1)
```

```python
elist['#'] = elist["#"].apply(lambda x:" ".join(x))
elist['#'] = elist["#"].apply(lambda x:x.lower())
```

```python
elist.head(1)
```

```python
vectors = cv.fit_transform(elist["#"]).toarray()
```

```python
cv.get_feature_names()
```

```python
elist["#"][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
def rec(movie):
    movie_index = elist[elist["title"] == movie].index[0]
    dist = similar[movie_index]
    movies_list = sorted(list(enumerate(dist)),reverse = True,key = lambda x:x[1])[0:6]
    for i in movies_list:
        print(elist.iloc[i[0]].title)
        
rec("Batman")
```
